```python
import pandas as pd

df_white = pd.read_csv("winequality-white.csv", sep=";")
df_red = pd.read_csv("winequality-red.csv", sep=";")
print(df_white.shape)
print(df_red.shape)
```

```python
df_white_copy = df_white.copy() 
df_red_copy = df_red.copy() 

df_white_copy['type'] = 'white'
df_red_copy['type'] = 'red'

# merging the datasets
data = pd.concat([df_white_copy,df_red_copy],ignore_index=True)
                 
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

#for i in data.columns:
#    sns.distplot(data[i])
#    plt.show()
```

```python
#data.type = data.type.map(dict(red=1, white=0))   # 1 if red, 0 if white

data = pd.get_dummies(data, columns=['type'])

print(data.groupby('quality').size())
```

```python
data
```

```python
data.describe()
```

```python
train = data.drop(['quality'], 1)    
test = data[['quality']]

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(train, test, train_size=0.7, random_state=42)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix

results = pd.DataFrame(columns = ['name', 'accuracy score', 'precision', 'recall']) 

RFC_n100 = RandomForestClassifier(n_estimators=100, random_state=42).fit(x_train, y_train.values.ravel())
RFC_n10 = RandomForestClassifier(n_estimators=10, random_state=42).fit(x_train, y_train.values.ravel())
RFC_n1000 = RandomForestClassifier(n_estimators=1000, random_state=42).fit(x_train, y_train.values.ravel())
DTC_gini = DecisionTreeClassifier(random_state=42, criterion='gini').fit(x_train, y_train)
DTC_entropy = DecisionTreeClassifier(random_state=42, criterion='entropy').fit(x_train, y_train)
DTC_entropy_minleaf1 = DecisionTreeClassifier(random_state=42, criterion='entropy', min_samples_leaf = 100).fit(x_train, y_train)
KNN_1 = KNeighborsClassifier(n_neighbors=1).fit(x_train, y_train.values.ravel())
KNN_5 = KNeighborsClassifier(n_neighbors=5).fit(x_train, y_train.values.ravel())
KNN_15 = KNeighborsClassifier(n_neighbors=15).fit(x_train, y_train.values.ravel())
KNN_50 = KNeighborsClassifier(n_neighbors=50).fit(x_train, y_train.values.ravel())
KNN_250 = KNeighborsClassifier(n_neighbors=250).fit(x_train, y_train.values.ravel())

from sklearn import preprocessing
x_train_processed = preprocessing.scale(x_train)
#x_test_processed = preprocessing.scale(x_test)

SVC_linear = SVC(kernel='linear').fit(x_train_processed, y_train.values.ravel())
SVC_poly = SVC(kernel='poly', degree=2, gamma='auto').fit(x_train_processed, y_train.values.ravel())
SVC_rbf =  SVC(kernel='rbf', gamma='auto').fit(x_train_processed, y_train.values.ravel())

models = [RFC_n100, RFC_n10, RFC_n1000, DTC_gini, DTC_entropy, DTC_entropy_minleaf1, KNN_1, KNN_5, KNN_15, KNN_50, 
          KNN_250, SVC_linear, SVC_poly, SVC_rbf]
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score, recall_score

#y_pred = pd.DataFrame(y_pred, columns=['prediction'])

for i in range(len(models)):
    y_pred = models[i].predict(x_test)
    results = results.append({'name' : str(models[i]), 'accuracy score' : accuracy_score(y_test, y_pred), 
                          'precision' : precision_score(y_test, y_pred, average='weighted', zero_division=1), 
                          'recall' : recall_score(y_test, y_pred, average='weighted', zero_division=1)},  
                         ignore_index = True)
```
